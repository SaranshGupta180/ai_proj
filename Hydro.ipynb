```python
# Gneiss:
# Drilling Technique: Air Rotary Drilling (DTH - Down the Hole Hammer)
# Reason: Gneiss is a hard, metamorphic rock, and air rotary with a hammer is effective for penetrating hard formations quickly and efficiently.


# Limestone:
# Drilling Technique: Mud Rotary or Air Rotary Drilling
# Reason: Limestone is often hard but may also contain fractures and voids. Mud rotary stabilizes the borehole in fractured zones, while air rotary works well in harder, more solid areas.

# Mica Schist:
# Drilling Technique: Rotary Drilling with Mud Circulation
# Reason: Schist is a medium-hard metamorphic rock with layers of mica that can break easily. Mud rotary prevents borehole collapse and supports the fragile layers during drilling.

# Older Alluvium:
# Drilling Technique: Mud Rotary Drilling
# Reason: Older alluvium is a mix of compacted sediments that can be unconsolidated. Mud rotary helps stabilize the borehole and prevents collapse in loose formations.

# Quartzite:
# Drilling Technique: Air Rotary Drilling (DTH)
# Reason: Quartzite is a very hard and durable metamorphic rock, requiring an air rotary technique with a hammer to penetrate efficiently.

# Quartzite Schist:
# Drilling Technique: Air Rotary or Mud Rotary Drilling
# Reason: Quartzite schist is hard like quartzite but also has schistose layers that can collapse. Air rotary works well for hard zones, while mud rotary stabilizes schistose layers.

# Schist:
# Drilling Technique: Mud Rotary Drilling
# Reason: Schist is layered and can be fragile. Mud rotary drilling supports the borehole, preventing it from collapsing due to the rock's weakness along its planes.

# Younger Alluvium:
# Drilling Technique: Mud Rotary Drilling
# Reason: Younger alluvium consists of unconsolidated sediments. Mud rotary drilling stabilizes the loose material and prevents collapse in soft, permeable zones.

# Yr. Alluvium (Younger Alluvium):
# Drilling Technique: Same as "Younger Alluvium" â€” Mud Rotary Drilling
# Reason: Same formation as mentioned above, requiring stabilization of loose sediment during drilling.
```

```python

x_hyd=hyd_for[["Latitude","Longitude"]]
y_hyd=hyd_for["Hyd_Formation"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
x_hyd_train, x_hyd_test, y_hyd_train, y_hyd_test = train_test_split(x_hyd, y_hyd, test_size=0.2, random_state=32)
model_hyd=KNeighborsClassifier(algorithm='auto',n_neighbors=9,weights='distance')
model_hyd.fit(x_hyd_train, y_hyd_train)
hyd_val=np.array([[26.6832448,75.2335695]])
hyd_pred=int(model_hyd.predict(hyd_val))
y_hyd_pred=model_hyd.predict(x_hyd_test)
print(hyd_pred)
accuracy = accuracy_score(y_hyd_test, y_hyd_pred)
print("Accuracy-> ",accuracy)
```
